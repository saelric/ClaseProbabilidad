```python
from sklearn import tree
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
model = tree.DecisionTreeClassifier()
```

```python
sirenas = pd.read_csv('sirenas_endemicas_y_sirenas_migrantes_historico.csv')
```

```python
y = pd.get_dummies(sirenas.pop("especie")) # Se utiliza "get_dummies" por tratarse de datos categoricos
x = sirenas
```

```python
x_train, x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state = 0)
```

```python
model = model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
sirenassinespecie = pd.read_csv('sirenas_endemicas_y_sirenas_migrantes.csv')
```

```python
y_sirenas = sirenassinespecie.pop("especie")
```

```python
y_sirenas = pd.DataFrame(data = model.predict(sirenassinespecie), columns = ['sirena_endemica','sirena_migrante'])
```

```python
y_sirenas["especie"] = np.where(y_sirenas["sirena_endemica"] == 1, 'sirena_endemica', 'sirena_migrante')
```

```python
sirenassinespecie["especie"] = y_sirenas.pop("especie")
```
